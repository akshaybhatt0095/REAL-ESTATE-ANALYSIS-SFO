```python
CSF=data[(data.city==' Fremont')|(data.city==' San Francisco')|(data.city==' San Mateo')|(data.city==' Berkeley')|(data.city==' San Jose')|(data.city==' Oakland')|(data.city==' Palo Alto')|(data.city==' Mountain View')|(data.city==' Piedmont')|(data.city==' Sunnyvale')|(data.city==' Half Moon Bay')|(data.city==' Walnut Creek')|(data.city==' San Ramon')|(data.city==' Millbrae')|(data.city==' Daly City')|(data.city==' Foster City')|(data.city==' Vallejo')|(data.city==' Pleasanton')|(data.city==' El Cerrito')|(data.city==' Menlo Park')]
CSF
```

```python

plt.figure(figsize=(10,5), dpi= 250)
sns.regplot('price','Price Variation Last 30 days',data=CSF,color='green',marker='+')
plt.title('CHANGES IN %s VERSUS %s' %('SOLD PRICE','PRICE VARIATION IN LAST 30 DAYS'))



```

```python
df_counts = CSF.groupby(['price', 'city']).size().reset_index(name='counts')

# Draw Stripplot
fig, ax = plt.subplots(figsize=(15,5), dpi= 400)    
sns.stripplot(df_counts.city, df_counts.price, size=df_counts.counts*10, ax=ax)

# Decorations
plt.title('Sold Price Scenario', fontsize=20)
plt.xticks(rotation=20)
plt.show()
```

```python
import random


# Prepare Data
df = CSF.groupby('city').size().reset_index(name='counts')
n = df['city'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

# Plot Bars
plt.figure(figsize=(16,8), dpi= 80)
plt.bar(df['city'], df['counts'], color=c, width=.5)
for i, val in enumerate(df['counts'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

# Decoration
plt.gca().set_xticklabels(df['city'], rotation=60, horizontalalignment= 'right')
plt.title("Number of properties sold", fontsize=22)
plt.ylabel('Count')
plt.ylim(0, 55)
plt.show()
```

```python
data=CSF
x_var = 'city'
groupby_var = 'bds'
df_agg = data.loc[:, [x_var, groupby_var]].groupby(groupby_var)
vals = [data[x_var].values.tolist() for i, data in df_agg]

plt.figure(figsize=(15,5), dpi= 100)
colors = [plt.cm.Spectral(i/float(len(vals)-1)) for i in range(len(vals))]
n, bins, patches = plt.hist(vals, data[x_var].unique().__len__(), stacked=True, density=False, color=colors[:len(vals)])

plt.legend({group:col for group, col in zip(np.unique(data[groupby_var]).tolist(), colors[:len(vals)])})
plt.title(f"Stacked Histogram of ${x_var}$ colored by ${groupby_var}$", fontsize=22)
plt.xlabel(x_var)
plt.ylabel("Frequency")
plt.ylim(0, 50)
plt.xticks(rotation=20)
plt.xticks(ticks=bins, labels=np.unique(data[x_var]).tolist(), rotation=90, horizontalalignment='left')
plt.show()
```
